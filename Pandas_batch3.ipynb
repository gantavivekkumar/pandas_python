```python
#Applying filters on data
#apply() method -->it takes function as an argument,and apply to all the values in the series/col
#df.apply(function,covert-dtype= True,args=(()))
d = {
    'name': ['a','b','c','d','e'],
    'marks':[56,70,40,80,90]

}
df = pd.DataFrame(d)
print(df)
#create a function
def funct(num):
  if num>60:
    return 'pass'
  else:
    return 'fail'
df['marks'].apply(funct,convert_dtype=True)
```

```python
#map() method
#by using map method it transforms values in series(col) using specified mapping data function
#maps takes dict,series/function
#new_sr = old.sr.map(mapping)

data = {"Grades":['A','B','C','D','E']}
df = pd.DataFrame(data)
print(df)
grades = {"A":"excellent","B":"good","C":"fair","D":"pass","E":"fail"}
df['Grades']= df['Grades'].map(grades)
print(df)
```

```python
import pandas as pd

# Define the dictionary and create the DataFrame
d = {
    'name': ['a', 'b', 'c', 'd', 'e'],
    'marks': [56, 70, 40, 80, 90],
    'Grades': ['A', 'B', 'C', 'D', 'E']  # Adding a Grades column
}
df = pd.DataFrame(d)

# Define the series to map grades
sr = pd.Series(['ex', 'Good', 'Fair', 'Pass', 'Fail'], index=['A', 'B', 'C', 'D', 'E'])
print("Series to map grades:")
print(sr)

# Map the 'Grades' column in df using the series sr
df['Grades'] = df['Grades'].map(sr)

# Print the DataFrame
print("DataFrame after mapping grades:")
print(df)
```

#Time series

```python
#What is time series data?
"""time series data is a sequence of data points in cronological order that is used by businessess
to analyze the past data and make the future predictions"""
#ex:- measuring the weather temparature
#no of taxi rides in a month
#predicting a company stock price for the next day
```

```python
#components of time series data.
# it consist of 4 componenets
#1. Trend component:- this is a variation that moves up or down in a reasonably predictable pattern over a long period
#2. Seasonality component:-the variation that is regular and periodicand repeats itself over a specific period
     #ex:- such as day, week,month,season, etc
#3.cyclic component:- is the variation that coressponds with business or economic "boom-bust" cycles or follows their own peculiar cycles
 #4.Random coponent or irregular coponent:-is the variation that is erratic or resedual and does not fall under any of the above three classification

```

#Groupby

```python
df = pd.DataFrame(
    {"A": ['one','one','two','three']*3,
     "B":['A','B','C',]*4,
     'C':['foo','foo','foo','bar','bar','bar']*2,
     'D':np.random.randn(12),
     'E':np.random.randn(12)}
)
```

```python
#group by column "c", "D","E"  .SUM()
```

```python
df.groupby('C')[['D','E']].sum()
#df.groupby('categorical column name')[[numerical column names]].aggrgate functions
```

```python
#grouping by multiple columns label forms multiindex
df.groupby(['A','C']).sum()
```

#merging or concat

```python
df = pd.DataFrame(np.random.randn(10,4))
#break it in to pieces
pieces = [df[:3],df[3:7],df[7:]]

pd.concat(pieces)
```

```python
#Merging of two data frames
left = pd.DataFrame({"key":['food','food'],'lval':[1,2]})
right = pd.DataFrame({"key":['food','food'],'rval':[3,4]})
#print left and right
# print(left)
# print(right)

pd.merge(left,right, on = 'key')
```
